```python
# kmeans_clustering.py

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Step 2: Prepare the Data
# Assuming you have a CSV file named 'customer_data.csv' with columns 'Annual Income' and 'Spending Score'
data = pd.read_csv('customer_data.csv')
X = data[['Annual Income', 'Spending Score']].values

# Step 3: Initialize and Fit the K-Means Model
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# Step 4: Obtain Cluster Assignments
cluster_labels = kmeans.labels_

# Step 5: Interpret Results
unique_labels, counts = np.unique(cluster_labels, return_counts=True)
for label, count in zip(unique_labels, counts):
    print("Cluster {}: {} data points".format(label, count))

# Step 6: Visualize Clusters
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering")
plt.show()
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Step 1: Read the Data and Explore It
df = pd.read_csv("C:\\Users\\Nukhba\\OneDrive - Punjab Group of Colleges\\Desktop\\datasets\\diabetes.csv" 
)
print(df.head())
print(df.describe())
print(df.isnull().sum())

# Step 2: Separate Labels from the Data
X = df.drop(columns=['Outcome'])
y = df['Outcome']

# Step 3: Implement KNN
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def knn(X_train, y_train, X_test, k):
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(X_train[i], X_test)
        distances.append((dist, y_train[i]))
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    output_values = [neighbor[1] for neighbor in neighbors]
    prediction = Counter(output_values).most_common(1)[0][0]
    return prediction

X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=42)

def predict(X_train, y_train, X_test, k):
    predictions = []
    for i in range(len(X_test)):
        prediction = knn(X_train, y_train, X_test[i], k)
        predictions.append(prediction)
    return predictions

# Step 4: Perform Prediction and Calculate Accuracy
k_values = [3, 5, 7, 11]
for k in k_values:
    y_pred = predict(X_train, y_train, X_test, k)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy for k={k}: {accuracy}')

# Step 5: Evaluate Your Model Using Different Evaluation Methods
def evaluate_model(y_test, y_pred):
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    return precision, recall, f1, cm

for k in k_values:
    y_pred = predict(X_train, y_train, X_test, k)
    precision, recall, f1, cm = evaluate_model(y_test, y_pred)
    print(f'Evaluation metrics for k={k}:')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-score: {f1}')
    print(f'Confusion Matrix:\n{cm}')
```
